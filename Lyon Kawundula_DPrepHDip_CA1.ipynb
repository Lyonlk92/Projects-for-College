# Cleaning and Analysing Data on Spam

- In this project I will be cleaning and analysing the data on emails sent to a company. 
- The data is labelled as true(is spam) and false(not spam)

```python
# dataset is imported using pandas as pd

import pandas as pd
```

```python
# dataset is imported and labeled df

df = pd.read_csv("spambase.csv")

# display data
df
```

This gives a summary of the data and we can see its made up of 4601 rows and 59 columns

```python

df.info()
```

- the code .info() is used to locate any Null data(noise)
- Overall the data in each column should total 4601
- A few of the columns do not have a total of 4601 such as world_freq_all
- This data will need to either be removed or replaced.
- Replacement cannot be accurate as data is a fact and we do not know if the data we provide is correct.


```python
# removing the null values from the columns

df = df.dropna(axis = 0, how ='any')

df.isna().sum()
```

```python
df.info()
```

- Rename the column - the first column is called unnammed.
- when plotting my graphs later i will not be able to discern what the data represents.
- by looking at the occurance of the data under unnamed:0, I can discern that this represents each email arriving,
- I can rename this column email_number

```python


# Rename the column
df_clean = df.rename(columns={'Unnamed: 0': 'email_count'})

df_clean
```

```python
# check if the date has any duplicates
# email count would be the correct variable to check

df_clean.duplicated("email_count").sum()
```

We can see our dataset has no duplicates
